# Data Loader

```python
lines = []
with open('03.txt', 'r') as f:
    lines = f.readlines()
line_arrays = [[int(x) for x in line.strip()] for line in lines]
```

# Part 1

```python
joltages = []
for line in line_arrays:
    start = max(line[0:-1])
    start_index = line.index(start)
    end=max(line[start_index+1:])
    joltage = int(str(start) + str(end))
    joltages.append(joltage)
total_joltage = sum(joltages)
total_joltage
```

# Part 2

```python
joltages = []
for line in line_arrays:
    numbers=[]
    start=0
    for i in range(12):
        m=max(line[start:len(line)-(11-i)])
        mpos=line[start:].index(m)+start
        numbers.append(str(m))
        start=mpos+1
    joltages.append(int(''.join(numbers)))

print(joltages,sum(joltages),sep='\n')
```
