# Data Loader

```python
topcount = 1000
with open("08.txt") as f:
    data =f.read().splitlines()
    points = [tuple(map(int, line.split(","))) for line in data]
# print(points)
```

# Part 1

```python
pair_distances = []
pairs = []
def maxi(l:list):
    max = 0
    mi = 0
    for i, v in enumerate(l):
        if v > max:
            max = v
            mi = i
    return mi
def distance(point1, point2):
    return ((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2 + (point1[2]-point2[2])**2)**0.5
for index, point in enumerate(points):
    for other_index in range(index+1, len(points)):
        other_point = points[other_index]
        pairs.append((point, other_point))
        pair_distances.append(distance(point, other_point))
sorted_pair_distances = sorted(pair_distances)[:topcount]
print("count of sorted pair distances", len(sorted_pair_distances))
print("sorted pair distances", sorted_pair_distances)
edges = []
for spair in sorted_pair_distances:
    i = pair_distances.index(spair)
    pair = pairs[i]
    edges.append(pair)
print(edges)
paths = []
while edges:    
    edge = edges[-1]
    foundPath = False
    for path in paths:
        if edge[0] in path or edge[1] in path :
            if edge[0] not in path:
                path.append(edge[0])
            if edge[1] not in path:
                path.append(edge[1])
            foundPath = True
    if not foundPath:
        paths.append(list(edge))
    edges.pop()


for index, path in enumerate(paths): 
    for point in path:
        doBreak = False
        while not doBreak:
            for other_index, other_path in enumerate(paths):
                if(other_index == paths.index(path)):
                    continue
                for other_point in other_path:
                    if(point == other_point):
                        # merge the two paths
                        paths.pop(other_index)
                        for other_point in other_path:
                            if(other_point not in path):
                                path.append(other_point)
                        break
            doBreak = True

print("paths", paths)


pathLens = [len(path) for path in paths]
pathLens.sort(reverse=True)
print("pathlens", pathLens)
topLens = pathLens[0:3]
print("top 3", topLens)

prod = 1
for path in topLens:
    prod *= path
print("product of top 3", prod)
```
