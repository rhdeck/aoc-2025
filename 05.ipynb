# Data Loader

```python
raw_data = open("05.txt").read().split("\n\n")
raw_IDs = raw_data[1].split("\n")
# print(raw_IDs[-1])
IDs = [int(id) if id != "" else None for id in raw_IDs]
raw_ranges = raw_data[0].split("\n")
_ranges = [range.split("-") for range in raw_ranges]
ranges = [[int(range[0]), int(range[1])] for range in _ranges]
# print(ranges,'\n',IDs)
```

# Part 1

```python
safes = []
for id in IDs:
    if not id:
        continue
    for range in ranges:
        if (id >= range[0] and id <= range[1]):
            safes.append(id)
            break
spoils=[x for x in IDs if x not in safes]
print(len(safes))
```

# Part 2

```python
sorted_ranges = sorted(ranges, key=lambda x: x[0])
fresh_ranges = []
for i, range in enumerate(sorted_ranges):
    if i == 0:
        fresh_ranges.append(range)
    elif range[0] <= sorted_ranges[i-1][1]:
        if range[1] > sorted_ranges[i-1][1]:
            fresh_ranges[-1][1] = range[1]
        else:
            pass
    else:
        fresh_ranges.append(range)
# print(fresh_ranges)
count=0
for range in fresh_ranges:
    count += range[1]-range[0]+1
print(count)
```
