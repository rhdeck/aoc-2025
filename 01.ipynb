```python
with open('01.test1.txt', 'r') as f:
    data = f.readlines()
    data = [line.strip() for line in data]

def parse_input(data):
    output = []
    for line in data:
        letter=line[0]
        number=int(line[1:])
        output.append((letter, number))
    return output
```

```python

pos = 50
zeroes=0
for letter, number in parse_input(data):
    sign= 1 if letter == 'R' else -1
    number=(sign*number) % 100
    pos = (number+pos) % 100
    if pos==0:
        zeroes+=1
print (pos, zeroes)





```
