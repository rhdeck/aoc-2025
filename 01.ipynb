# Data Loader

```python
with open('01.txt', 'r') as f:
    data = f.readlines()
    data = [line.strip() for line in data]

def parse_input(data):
    output = []
    for line in data:
        letter=line[0]
        number=int(line[1:])
        output.append((letter, number))
    return output
```

# Part 1

```python

pos = 50
zeroes=0
for letter, number in parse_input(data):
    sign= 1 if letter == 'R' else -1
    number=(sign*number) % 100
    pos = (number+pos) % 100
    if pos==0:
        zeroes+=1
print ("Part 1", zeroes)





```

# Part 2

```python
pos = 50
zeroes=0
for letter, number in parse_input(data):
    # zero_clicked=False
    sign= 1 if letter == 'R' else -1
    rawNumber=sign*number
    test = rawNumber + pos
    if test <= 0:
        zero = abs(test) // 100 + (1 if pos > 0 else 0)
        if zero > 0:
            zeroes+= zero
    # else:
    elif test > 99:
        zeroes+= abs(test) // 100 
    number=(sign*number) % 100
    pos = (number+pos) % 100
print ("Part 2", zeroes)
```
