# Data Loader

```python
import numpy as np
with open('07.txt') as f:
    lines = f.readlines()
    source=lines[0]
    grid=lines[1:]
    lines = [list(line.strip()) for line in grid]
    grid = np.array(lines)
    start = source.index('S')
```

# Part 1

```python
splits = 0
current = [False] * len(source)
future = [False] * len(source)
current[start] = True
for line in grid:
    for i, c in enumerate(line):
        if c == '^' and current[i]:
            splits +=1
            future [i-1] = True
            future [i+1] = True
        elif current[i] and c == '.':
            future[i] = True
    current = future
    future = [False] * len(source)

print(splits)
```

# Part 2 - Dynamic Programmming

```python
timeweights = [0] * len(source)
timeweights[start] = 1
for line in grid:
    for i, c in enumerate(line):
        if c == "^" and timeweights[i] != 0:
            timeweights[i-1] += timeweights[i]
            timeweights[i+1] += timeweights[i]
            timeweights[i] = 0
print(sum(timeweights))
```
