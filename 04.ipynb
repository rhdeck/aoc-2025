# Part 1

```python
def get_neighbors(grid): 
    neighbors = []
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j]:
                # get neighbors
                neighbor_count = 0 
                for row_offset in [-1, 0, 1]:
                    for col_offset in [-1, 0, 1]:
                        if row_offset == 0 and col_offset == 0:
                            continue
                        neighbor_row = i + row_offset
                        neighbor_col = j + col_offset
                        if(neighbor_row > -1 and neighbor_row < len(grid) and neighbor_col > -1 and neighbor_col < len(grid[0])):
                            if(grid[neighbor_row][neighbor_col]):
                                neighbor_count += 1
                if(neighbor_count < 4):
                    neighbors.append((i, j))
    return neighbors
neighbors = get_neighbors(grid)
print(len(neighbors))
```

# Part 2

```python
from IPython.display import clear_output
from time import sleep
delay = .05
g = []
with open('04.txt', 'r') as f:
    for line in f:
        g.append([c == '@' for c in line.strip()])
removals = []
for counter in range(10000000):
    neighbors = get_neighbors(grid)
    if(len(neighbors) == 0):
        break
    removals.append(len(neighbors))
    for i, j in neighbors:
        grid[i][j] = False
    clear_output(wait=True)
    sleep(delay)
    for row in grid:
        print(''.join(['#' if c else '.' for c in row]))
    sleep(delay)

grid = g
    

# print(removals)

sum(removals)

```
