# Data Loader

```python
elements = open("02.txt").read().split(",")

# Parse the elements into a list of tuples
input = [tuple(map(int, element.split("-"))) for element in elements]

input
```

# Part 1

```python
invalids = []
for (start, end) in input:
    for n in range(int(start), int(end)+1):
        s = str(n)
        if len(s) % 2 == 1:
            continue
        # print(s, s[:(len(s)//2)], s[(len(s)//2):])
        if s[:(len(s)//2)] == s[(len(s)//2):]:   
            invalids.append(n)
print(sum(invalids),'\n',invalids)
```

# Part 2

```python
invalids = []
for (start, end) in input:

    for n in range(int(start), int(end)+1):
        s = str(n)
        # print(s)
        l=len(s)
        for i in range(1,l//2+1):
            # print('len',l,i)
            if l%i>0:
                continue
            c=s[0:i]
            is_match = True
            for j in range(l//i):
                
                d = s[j*i:j*i+i]
                # print(s, d,c, d==c, j)
                if d != c:
                    is_match = False
                    break
            if is_match:
                invalids.append(n)
                break
print(sum(invalids))
invalids
```
