# Data Loader - using Pandas

```python
import pandas as pd
import re

# Split each line on any whitespace to get individual values
# This treats each number as a separate column
with open('06.txt') as f:
    data = []
    for line in f:
        line = line.strip()
        if line:
            # Split on any whitespace (handles single and multiple spaces)
            values = re.split(r'\s+', line)
            data.append(values)

# Find maximum number of columns
max_cols = max(len(row) for row in data)

# Pad rows to same length with None/NaN
for row in data:
    while len(row) < max_cols:
        row.append(None)

# Create dataframe
df = pd.DataFrame(data)
display(df.head())

```

# Part 1

```python
totals = []
def prod(l:list):
    total =1 
    for i in l:
        total *= i
    return total

for col_name, col_data in df.items():
    # print(col_data)
    col_numbers = pd.to_numeric(col_data, errors='coerce').dropna().tolist()
    # print(col_numbers)
    if(col_data.iloc[-1] == '+'):
        totals.append(sum(col_numbers))
    else:
        totals.append(prod(col_numbers))

print(totals)

int(sum(totals))


```

```python
import numpy as   np
with open('06.txt') as f:
    lines = f.readlines()
    operations = lines[-1]
    indexes = [i for i, c in enumerate(operations) if (c == '*' or c == '+')]
    operators = [c for c in operations if c == '*' or c == '+']
    listsOfNumbers = [[] for _ in range(len(indexes))]
    for line in lines[:-1]:
        for array_index, position in enumerate(indexes):
            after_position = indexes[array_index + 1] if array_index < len(indexes) - 1 else len(line)
            value = line[position:after_position-1]
            listsOfNumbers[array_index].append(np.array([c for c in value]))
    listsOfRealNumbers = []
    for listOfNumbers in listsOfNumbers:
        listOfRealNumbers = []
        #this is where my boy transposes the list
        listOfNumbers = np.array(listOfNumbers).T
        for line in listOfNumbers:
            line = int(''.join(line))
            listOfRealNumbers.append(line)
        listsOfRealNumbers.append(listOfRealNumbers)
    totals = []
    for index, numberList in enumerate(listsOfRealNumbers):
        if operators[index] == '+':
            totals.append(sum(numberList))
        else:
            totals.append(prod(numberList))
    print(totals)
    print(sum(totals))
```
